### We demonstrate how to apply hierarchical clustering algorithm for data discovery

```python
import pandas as pd
import numpy as np
from sklearn import datasets
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
# import Iris dataset
iris = datasets.load_iris()
data = pd.DataFrame(iris.data)
data_scaled = normalize(data)
```

```python
# Hierarchical clustering 
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend_single = shc.dendrogram(shc.linkage(data, method='single'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
cluster.fit_predict(data)
```

```python
#avg
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend_avg = shc.dendrogram(shc.linkage(data, method='average'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')
cluster.fit_predict(data)
```

### We demonstrate how to apply k-means clustering algorithm for data discovery

```python
#k-means
from sklearn.cluster import KMeans

data = [2, 4, 10, 12, 3, 20, 30, 11, 25]
data = np.array(data)
initial = np.array([[10],[15]], np.float64)
kmeans = KMeans(n_clusters=2, init = initial,n_init=1).fit(data.reshape(-1,1))
labels = kmeans.labels_
```

```python
filtered_label0 = data[labels == 0]
filtered_label1 = data[labels == 1]

plt.scatter(filtered_label0 , [0]*len(filtered_label0), color = 'red')
plt.scatter(filtered_label1 ,[0]*len(filtered_label1), color = 'black')
plt.show()
```

```python
# k-means for iris dataset
# import Iris dataset

import mpl_toolkits.mplot3d  # noqa: F401
iris = datasets.load_iris()

X = iris.data
y = iris.target

# Kmeans with different initializations
estimators = [
    ("k_means_iris_8", KMeans(n_clusters=8)),
    ("k_means_iris_3", KMeans(n_clusters=3)),
    ("k_means_iris_bad_init", KMeans(n_clusters=3, n_init=1, init="random")),
]

fignum = 1

titles = ["8 clusters", "3 clusters", "3 clusters, bad initialization"]
```

```python
X
```

```python
for name, est in estimators:
    fig = plt.figure(fignum, figsize=(4, 3))
    ax = fig.add_subplot(111, projection="3d", elev=48, azim=134)
    ax.set_position([0, 0, 0.95, 1])
    est.fit(X)
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=labels.astype(float), edgecolor="k")

    ax.w_xaxis.set_ticklabels([])
    ax.w_yaxis.set_ticklabels([])
    ax.w_zaxis.set_ticklabels([])
    ax.set_xlabel("Petal width")
    ax.set_ylabel("Sepal length")
    ax.set_zlabel("Petal length")
    ax.set_title(titles[fignum - 1])
    ax.dist = 12
    fignum = fignum + 1

# Plot the ground truth
fig = plt.figure(fignum, figsize=(4, 3))
ax = fig.add_subplot(111, projection="3d", elev=48, azim=134)
ax.set_position([0, 0, 0.95, 1])

for name, label in [("Setosa", 0), ("Versicolour", 1), ("Virginica", 2)]:
    ax.text3D(
        X[y == label, 3].mean(),
        X[y == label, 0].mean(),
        X[y == label, 2].mean() + 2,
        name,
        horizontalalignment="center",
        bbox=dict(alpha=0.2, edgecolor="w", facecolor="w"),
    )
# Reorder the labels to have colors matching the cluster results
y = np.choose(y, [1, 2, 0]).astype(float)
ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=y, edgecolor="k")

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
ax.set_title("Ground Truth")
ax.dist = 12

fig.show()
```
