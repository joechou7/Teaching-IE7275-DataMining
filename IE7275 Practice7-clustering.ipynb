### We demonstrate how to apply hierarchical clustering algorithm for data discovery

```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np
from sklearn import datasets
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
```

```python
# import Iris dataset
iris = datasets.load_iris()
data = pd.DataFrame(iris.data)
data_scaled = normalize(data)
```

```python
# Hierarchical clustering 
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend_single = shc.dendrogram(shc.linkage(data, method='single'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='single')
cluster.fit_predict(data)
```

```python
#avg
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend_avg = shc.dendrogram(shc.linkage(data, method='average'))
plt.axhline(y=6, color='r', linestyle='--')
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')
cluster.fit_predict(data)
```

### We demonstrate how to apply k-means clustering algorithm for data discovery

```python
#k-means
data = [2, 4, 10, 12, 3, 20, 30, 11, 25]
data = np.array(data)
initial = np.array([[10],[15]], np.float64)
kmeans = KMeans(n_clusters=2, init = initial,n_init=1).fit(data.reshape(-1,1))
labels = kmeans.labels_
```

```python
filtered_label0 = data[labels == 0]
filtered_label1 = data[labels == 1]

plt.scatter(filtered_label0 , [0]*len(filtered_label0), color = 'red')
plt.scatter(filtered_label1 ,[0]*len(filtered_label1), color = 'black')
plt.show()
```
